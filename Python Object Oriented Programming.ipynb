### Inheritance
Inheritance in Python

Inheritance is a powerful feature in object oriented programming.

It refers to defining a new class with little or no modification to an existing class. The new class is called derived (or child) class and the one from which it inherits is called the base (or parent) class.



Python Inheritance Syntax



```python
class BaseClass:
  #Body of base class
class DerivedClass(BaseClass):
  #Body of derived class
```

```python
class parent:
    data1="parent data"
    def properties1(self):
        print("Inside parent class")
        
class child(parent):  # class classname(parent_class):
    data2="child data"
    def properties2(self):
        print("Inside child class")
```

```python
p=parent()
print(p.data1)
p.properties1()
```

```python
c=child()
print(c.data2)
c.properties2()
print(c.data1)
c.properties1()
```

```python
class parent1:
    data1="parent1 data"
    def properties1(self):
        print("Inside parent1 class")
        
class parent2:  
    data2="parent2 data"
    def properties2(self):
        print("Inside parent2 class")
        
class child(parent1,parent2):  
    data3="child data"
    def properties3(self):
        print("Inside child class")
```

```python
c=child()
print(c.data2)
c.properties2()
print(c.data1)
c.properties1()
print(c.data3)
c.properties3()
```

```python
class Parent1:
    data1 = "parent1 data"
    def properties1(self):
        print("Inside parent1 data")
class child1(Parent1):
    data2 = "child1 data"
    def properties2(self):
        print("Inside child1 data")
class child2(child1):
    data3 = "child2 data"
    def properties3(self):
        print("Inside child2 data")
```

```python
class Parent1:
    data1 = "parent1 data"
    def properties1(self):
        print("Inside parent1 data")
class child1(Parent1):
    data2 = "child1 data"
    def properties2(self):
        print("Inside child1 data")
class child2(Parent1):
    data3 = "child2 data"
    def properties3(self):
        print("Inside child2 data")
```

### Hybrid Inheritance


```python
class Parent1:
    data1 = "parent1 data"
    def properties1(self):
        print("Inside parent1 data")
class child1(Parent1):
    data2 = "child1 data"
    def properties2(self):
        print("Inside child1 data")
class child2(Parent1):
    data3 = "child2 data"
    def properties3(self):
        print("Inside child2 data")
class hybridChild(child1,child2):
    data3 = "hybrid Child data"
    def properties3(self):
        print("Inside hybrid Child data")
```
